```python
import matplotlib.pyplot as plt
import numpy as np
xpts = np.array([100, 250 ])
ypts = np.array([200, 400])
plt.xlabel("Chocolate rate")
plt.ylabel("Choco grams")
plt.plot(xpts, ypts)
plt.grid()
plt.title('Choco Rates')
plt.show()
```

```python
Sick_leave = np.array([4, 12, 2, 25]) 
plt.xlabel("no of days")
plt.ylabel("Diff of 5 days")
plt.title('Sick days')
plt.grid()
plt.plot(Sick_leave, marker='s') #add square markers at data points
plt.show()
```

```python
plt.plot(Sick_leave,linestyle='dotted') #unless specifid x = 0,1,2...
plt.show()
```

```python
#subplots
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([2000, 2010, 2020, 2030])
y1 = np.array([6, 3, 12, 10])
plt.subplot(1, 2, 1)
plt.plot(x1,y1)
x2 = np.array([2000, 2010, 2020, 2030])
y2 = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(x2,y2)
plt.show()
```

```python
#Scatter Plots
A = np.array([2, 3, 4, 11, 12, 17, 22, 39, 14, 21, 23, 9, 6])
B = np.array([59, 26, 67, 78, 121, 23, 20, 69, 93, 45, 24, 1, 5])
print("len A = ",len(A))
print("len B = ",len(B))
plt.scatter(A,B)
plt.show()
```

```python
#Bar plot
plt.bar(A,B)
plt.show()
```

```python
A = np.random.normal(134, 20, 450)
plt.hist(A) #histogram give frequecy of reoccuring events
plt.show()
```
