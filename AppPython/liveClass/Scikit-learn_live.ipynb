```python
#standarization: mean = 0, sigma = 1
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'x':np.random.normal(0,3,10000), #(mean,sigma,population size)
    'y':np.random.normal(6,4,10000),
    'z':np.random.normal(-6,6,10000)
    })
df.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
standardscalar = StandardScaler()
data_tf = standardscalar.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x', 'y', 'z'])
df.plot.kde()
```

```python
df = pd.DataFrame({
    'x':np.random.chisquare(8, 1000), #pos skew
    'y':np.random.beta(8, 2, 1000)*40, #neg skew
    'z':np.random.normal(50, 3, 1000) #no skew
})
df.plot.kde()
```

```python
from sklearn.preprocessing import StandardScaler
standardscalar = StandardScaler()
data_tf = standardscalar.fit_transform(df)
df = pd.DataFrame(data_tf, columns=['x', 'y', 'z'])
df.plot.kde()
```
