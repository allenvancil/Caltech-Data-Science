```python
 #1. import libraries
 import numpy as np
 import matplotlib.pyplot as plt
 from matplotlib import style
 import pandas as pd
 %matplotlib inline
```

```python
#2. import/create data set
randomnumber = np.random.rand(10)
print([randomnumber])
```

```python
#3. set parameters
style.use('ggplot')
plt.plot(randomnumber, 'g', label='line one', linewidth=2)
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
```

```python
#make pie chart, define data, libs imported above
cause = 'Chronic Disease', 'Unintentional Injuries', 'Alzheimers', 'Influenza and Pneumonia', 'Sepsis', 'Other'
percentile = [62, 5, 4, 2, 1, 26]
```

```python
#Set parameters
plt.figure(figsize=(10,10)) #size of plot
explode = (0.05, 0, 0, 0, 0, .1) #breaks element for chart by specified amount
plt.pie(percentile, labels=cause, explode=explode, autopct='%1.1f%%', startangle=70)
        #
plt.axis('equal')
plt.title('Ohio State - 2012 : Leading Causes of Death')
plt.show()
```
